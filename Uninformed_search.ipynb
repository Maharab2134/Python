```python
graph = {
    'A' : ['B', 'C'],
    'B' : ['D', 'M'],
    'C' : ['L', 'P'],
    'D' : [],
    'M' : ['K'],
    'L' : ['R'],
    'P' : ['Y'],
    'K' : [],
    'R' : ['G'],
    'Y' : [],
}
```

```python
# BFS
from collections import deque

def BFS(graph, start):
    visited = set()
    queue = deque([start])

    while queue:
        node = queue.popleft()
        if node not in visited:
            print(node, end=" ")
            visited.add(node)
            queue.extend(graph.get(node, []))


BFS(graph, 'A')
```

```python
# DFS Left
from collections import deque

def DFS(graph, start):
    visited = set()
    stack = deque([start])

    while stack:
        node = stack.pop()
        if node not in visited:
            print(node, end=" ")
            visited.add(node)
            stack.extend(graph.get(node, []))

DFS(graph, 'A')
```

```python
# DFS Right
from collections import deque

def DFS(graph, start):
    visited = set()
    stack = deque([start])

    while stack:
        node = stack.pop()
        if node not in visited:
            print(node, end=" ")
            visited.add(node)
            stack.extend(reversed(graph.get(node, [])))
DFS(graph, 'A')
```

```python
# DLS Left
from collections import deque

def DLS(graph, start, dep_l):
    visited = set()
    stack = deque([(start, 0)])

    while stack:
        node, depth = stack.pop()

        if depth > dep_l: ##
            continue

        if node not in visited:
            print(node, end=" ")
            visited.add(node)

            for neighbor in graph.get(node, []):
                stack.append((neighbor, depth + 1))

DLS(graph, 'A', 2)
```

```python
# DLS Right

from collections import deque

def DLS(graph, start, dep_l):
    visited = set()
    stack = deque([(start, 0)])

    while stack:
        node, depth = stack.pop()
        if depth > dep_l:
            continue

        if node not in visited:
            print(node, end=" ")
            visited.add(node)

            for neighbor in graph.get(node, []):
                stack.append((neighbor, depth + 1))
                stack.reverse()
DLS(graph, 'A', 2)
```

```python
# Goal BFS
graph = {
    'A' : ['L', 'M', 'C'],
    'L' : ['P'],
    'M' : ['K', 'R'],
    'C' : ['T', 'G'],
    'P' : ['L'],
    'K' : ['M'],
    'R' : ['M'],
    'T' : ['C'],
    'G' : ['C']
}
from collections import deque

def bfs(graph, start, goal):
    queue = deque([(start, [start])])
    visited = set()

    while queue:
        node, path = queue.popleft()

        if node == goal:
            return path

        visited.add(node)

        for neighbor in graph.get(node, []):
            if neighbor not in visited:
                queue.append((neighbor, path + [neighbor]))

    return None

result = bfs(graph, 'A', 'T')
print("Goal:", result)
```

```python
# Goal DFS (Left)

from collections import deque

def dfs(graph, start, goal):
    stack = deque([start])
    visited = set()
    path = []

    while stack:
        node = stack.pop()

        if node == goal:
            path.append(node)
            return path

        if node not in visited:
            visited.add(node)
            path.append(node)
            for neighbor in reversed(graph.get(node, [])):
                if neighbor not in visited:
                    stack.append(neighbor)

    return None

result = bfs(graph, 'A', 'T')
print("Path to goal:", result)

```

```python
# Goal DFS (Right)

from collections import deque

def dfs(graph, start, goal):
    stack = deque([start])
    visited = set()
    path = []

    while stack:
        node = stack.pop()

        if node == goal:
            path.append(node)
            return path

        if node not in visited:
            visited.add(node)
            path.append(node)
            for neighbor in graph.get(node, []):
                if neighbor not in visited:
                    stack.append(neighbor)

    return None

result = dfs(graph, 'A', 'T')
print("Goal:", result)
```
